```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("x-y")
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y)
ax2.plot(x,y)
fig
```
